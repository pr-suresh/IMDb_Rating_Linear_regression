```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv("newmoviesall.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
all_columns = list(df) # Creates list of all column headers
df[all_columns] = df[all_columns].astype(str)
```

```python
df.info()
```

```python
df
```

```python
df['country_of_origin'] = df['country_of_origin'].map(lambda x: x.lstrip('Country of origin'))
```

```python
df.loc[df['country_of_origin'] == 'United States']
```

```python
#df['popularity'] = df['popularity'].astype(str)
df["popularity"] = df["popularity"].map(lambda x: x.replace(',',''))
```

```python
def movie_length(movie):
    word_list = movie.split()
    return len(word_list)
```

```python
df["movie_length"] = df["movie"].apply(movie_length)
```

```python
def camel_case_split(str):
    str1 =re.findall(r'[A-Z](?:[a-z]+|[A-Z]*(?=[A-Z]|$))', str)
    return str1[0]
```

```python
df['genre'] = df['genre'].apply(camel_case_split)
df['imdb'] = df['imdb'].map(lambda x: x.strip('IMDb RATING'))
```

```python
df[['imdb', 'votes']] = df.imdb.str.split('/', expand=True)
```

```python
df['votes'] = df['votes'].map(lambda x: x[2:] if x is not None else np.nan)
```

```python
def runtime_cal(str1):
    hrs = str1[:1]
    mins = str1[3:5]
    try:
        minutes = int(hrs)*60 + int(mins)
        return minutes
    except:
        return None
```

```python
df['runtime'] = df['runtime'].map(lambda x: x.lstrip('Runtime'))
df['runtime'] = df['runtime'].map(runtime_cal)
```

```python
df.loc[df['country_of_origin'] == 'United States']
```

```python
df['country_of_origin'] = df['country_of_origin'].map(lambda x: x.lstrip('Country of origin'))
#df['country_of_origin'] = df['country_of_origin'].apply(lambda x: 'United States' if 'United States' in x else x)
```

```python
df['director'] = df['director'].map(lambda x: x.lstrip('Director'))
```

```python
df['writer'] = df['writer'].map(lambda x:np.nan if x== "['']" else x)
df['actor'] = df['actor'].map(lambda x:np.nan if x== "['']" else x)
```

```python
df.loc[df['country_of_origin'] == 'United States']
```

```python
def set_ratings(mpaa):
    ratings = ['PG-13','TV-MA', 'PG', 'G', 'TV-G', 'TV-Y', 'R']
    for rating in ratings:
        if rating in mpaa:
            return rating
    return 'Not Rated'
```

```python
df['mpaa'] = df['mpaa'].map(set_ratings)
```

```python
df['mpaa'].unique()
```

```python
df['budget'] = df['budget'].map(lambda x: x.strip(' (estimated) '))
df['budget'] = df['budget'].map(lambda x: x.replace('$', '').replace(',', ''))
```

```python
df["box_office_dom"] = df['box_office_dom'].map(lambda x: x.strip(' (estimated) '))
df['box_office_dom'] = df['box_office_dom'].map(lambda x: x.replace('$', '').replace(',', ''))
```

```python
df["box_office_wknd1"] = df['box_office_wknd1'].map(lambda x: x.strip(' (estimated) '))
df['box_office_wknd1'] = df['box_office_wknd1'].map(lambda x: x.replace('$', '').replace(',', ''))
```

```python
df["box_office_gross"] = df['box_office_gross'].map(lambda x: x.strip(' (estimated) '))
df['box_office_gross'] = df['box_office_gross'].map(lambda x: x.replace('$', '').replace(',', ''))
```

```python
def convert_str_to_number(x):
    tot_votes = 0
    if 'K' in str(x):
        x=float(x.replace('K',''))
        tot_votes = x*1000
        return int(tot_votes)
    elif x is np.nan:
        return np.nan
    else:
        return int(float(x))
```

```python
df['votes']= df['votes'].map(convert_str_to_number)
```

```python
df = df.replace('nan', np.nan)
```

```python
df.dropna()
```

```python
df2 = df.loc[df['country_of_origin'] == 'United States'].copy()
```

```python
df2.drop_duplicates(inplace=True)
```

```python
df2.drop(['popularity', 'metascore'], axis=1, inplace=True)
```

```python
df2.dropna()
```

```python
all_columns = ['budget', 'box_office_dom', 'box_office_wknd1', 'box_office_gross', 'metascore', 'popularity']
df2[all_columns] = df2[all_columns].astype(float)
```

```python
df2.to_csv("IMDB_movies.csv")
```
